```python
def text_cleaner(text):
    text = re.sub("[\[].*?[\]]", "", text)
    
    # Get rid of extra whitespace.
    text = ' '.join(text.split())
    
    return text
```

```python
def word_extraction(text):    
    ignore = stopwords    
    words = re.sub("[^\w]", " ",text).split()    
    clean1text = [w.lower()
    for w in words
        if w not in ignore]
    
    return clean1text
```

```python
from collections import Counter

# Utility function to calculate how frequently words appear in the text.
def word_frequencies(text, include_stop=False):
    
    # Build a list of words.
    # Strip out punctuation and, optionally, stop words.
    words = []
    for token in text:
        if not token.is_punct and (not token.is_stop or include_stop):
            words.append(token.text)
            
    # Build and return a Counter object containing word counts.
    return Counter(words)
    

    
    # The most frequent words:


df2015nlp_freq = word_frequencies(df2015nlp, include_stop=False).most_common(20)
df2016nlp_freq = word_frequencies(df2016nlp, include_stop=False).most_common(20)
df2017nlp_freq = word_frequencies(df2017nlp, include_stop=False).most_common(20)
df2018nlp_freq = word_frequencies(df2018nlp, include_stop=False).most_common(20)
print('2015:', df2015nlp_freq)
print('2016:', df2016nlp_freq)
print('2017:', df2017nlp_freq)
print('2018:', df2018nlp_freq)
```
